### Plot Scatterplot and Countplot Using Seaborn

```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.head()
```

```python
# Plot scatter plot between mean area and mean smoothness using hue as target
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = cancer_df);
```

```python
# Plot scatter plot between mean area and mean smoothness
sns.scatterplot(x = 'mean area', y = 'mean smoothness',data = cancer_df)
```

```python
# Countplot to know how many samples belong to class #0 and #1
sns.countplot(x = cancer_df['target']) 
```

### Pairplot, Heatmap, Distplot

```python
# Plot the pairplot
sns.pairplot(cancer_df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```

```python
# Heatmap to find correlation
plt.figure(figsize = (20, 10)) 
sns.heatmap(cancer_df.corr(), annot = True) 
```

```python
# Distplot 
sns.distplot(x = cancer_df['mean radius'], bins = 15, color = 'blue')
```

###  Plot two separate distplot for each target class 0 and 1

```python
class_0_df = cancer_df[cancer_df['target'] == 0]
class_0_df
```

```python
class_1_df = cancer_df[cancer_df['target'] == 1]
```

```python
plt.figure(figsize = (10, 7))
sns.distplot(class_0_df['mean radius'], bins = 15, color = 'blue')
sns.distplot(class_1_df['mean radius'], bins = 15, color = 'red')
```
