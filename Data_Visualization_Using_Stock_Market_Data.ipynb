# Data Visualization Using Matplotlib and Seaborn

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Import datasets using Pandas.
stocks_prices_df = pd.read_csv('stocks_daily_prices.csv')
stocks_prices_df
```

```python
# Read daily return data using pandas
daily_returns_df = pd.read_csv('stocks_daily_returns.csv')
daily_returns_df
```

```python
# Plot lineplots to display all 4 stocks daily prices on one single figure
stocks_prices_df.plot(x = 'Date', y = ['AAPL', 'JPM', 'PG', 'UAL'], linewidth = 3, figsize = (16, 8))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc = "upper center")
plt.grid()
```

```python
# plot 4 stocks daily prices on multiple subplots
stocks_prices_df.plot(x = 'Date', y = ['AAPL', 'JPM', 'PG', 'UAL'], subplots = True, grid = True, linewidth = 3, figsize = (16, 16))
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend(loc = "upper center")
```

```python
# Plot the scatterplot between Apple and JP Morgan daily returns
X = daily_returns_df['AAPL']
Y = daily_returns_df['JPM']
plt.figure(figsize = (15, 10))
plt.scatter(X, Y);
plt.grid()
```

```python
# Plot similar scatterplot between Apple and JP Morgan daily returns
plt.figure(figsize = (15, 8))
sns.scatterplot(x = 'AAPL', y = 'JPM', data = daily_returns_df);
plt.grid()
```

```python
# Plot a pie chart that shows these allocations 
# Use 'explode’ attribute to increase the separation between AAPL and the rest of the portfolio
values  = [10, 70, 10, 10]
colors  = ['g', 'r', 'y', 'b']
explode = [0, 0.2, 0, 0]
labels  = ['JP Morgan Chase (JPM)', 'APPLE (AAPL)', 'United Airlines (UAL)', 'P&G (PG)']

# Use matplotlib to plot a pie chart 
plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('STOCK PORTFOLIO')
plt.show()
```

```python
# Plot the histogram for United Airlines and P&G returns using 40 bins with red color
# Display the mean and Standard deviation for both stocks on top of the figure

# Plot the histogram of the returns
plt.figure(figsize = (20, 15)); 

mu_UAL = round(daily_returns_df['UAL'].mean(), 2)
sigma_UAL = round(daily_returns_df['UAL'].std(), 2)
    
mu_PG = round(daily_returns_df['PG'].mean(), 2)
sigma_PG = round(daily_returns_df['PG'].std(), 2)
    
daily_returns_df[['PG','UAL']].plot.hist(bins = 60, alpha = 0.7, figsize = (15, 8));
plt.grid();
plt.ylabel('Probability')


plt.title('Histogram: P&G mu=' + str(mu_PG) + ', P&G sigma=' +str(sigma_PG) + ', United Airlines mu=' +str(mu_UAL) + ', United Airlines sigma=' +str(sigma_UAL) );
```

```python
# Plot a heatmap that shows the correlations between stocks daily returns
plt.figure(figsize = (12, 10))
cm = daily_returns_df.corr()
sns.heatmap(cm, annot = True)
```

```python
# Plot a 3D plot showing all daily returns from JPM, AAPL and UAL
# Toolkits are collections of application-specific functions that extend Matplotlib
# mpl_toolkits.mplot3d provides tools for basic 3D plotting.
# https://matplotlib.org/mpl_toolkits/index.html

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection = '3d')

x = daily_returns_df['JPM'].tolist()
y = daily_returns_df['AAPL'].tolist()
z = daily_returns_df['UAL'].tolist()

ax.scatter(x, y, z, c = 'r', marker = 'o', s = 1000)

ax.set_xlabel('JP Morgan')
ax.set_ylabel('Apple')
ax.set_zlabel('United Airlines')
```
