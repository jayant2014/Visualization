```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Read crypto daily prices data and convert into pandas dataframe
crypto_df = pd.read_csv('crypto_daily_prices.csv')
crypto_df
```

```python
# Use matplotlib to plot the data
crypto_df.plot(x = 'Date', y = 'BTC', label = 'Bitcoin Price', color = 'r', linewidth = 3, figsize = (14, 6));
plt.ylabel('Price [$]')
plt.title('Data Visualization of Crypto!')
plt.legend(loc = "upper right")
plt.grid()
```

```python
# Let's plot multiple cryptos in a single plot
crypto_df.plot(x = 'Date', y = [ 'BTC', 'ETH', 'ADA'], linewidth = 3, figsize = (14,6))
plt.ylabel('Price')
plt.title('Crypto Prices')
plt.grid()
```

```python
# Use subplots attribute to plot separate figures for BTC, ETH, and ADA
crypto_df.plot(x = 'Date',title = 'Crypto Prices', subplots = True, figsize = (14,6))
```

### Plot Scatterplots 

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
# Plot Daily returns of BTC vs. ETH
daily_return_df.plot.scatter('BTC', 'ETH', grid = True, figsize = (12, 7), color = 'blue', alpha = 0.5);
```

```python
# Plot Daily returns of BTC vs. ADA
daily_return_df.plot.scatter('BTC', 'ADA', grid = True, figsize = (12, 7), color = 'red', alpha = 0.5);
```

### Plot Histogram

```python
# Mean of distribution
mu = round(daily_return_df['BTC'].mean(), 2)
# S deviation of distribution
sigma = round(daily_return_df['BTC'].std(), 2) 
 
num_bins = 40
plt.figure(figsize = (20, 15)) 

daily_return_df['BTC'].plot.hist(bins = num_bins, color = 'b', alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));
```

```python
# Mean of distribution
mu = round(daily_return_df['ETH'].mean(), 2)
# S deviation of distribution
sigma = round(daily_return_df['ETH'].std(), 2) 
 
num_bins = 40
plt.figure(figsize = (20, 15)) 

daily_return_df['ETH'].plot.hist(bins = num_bins, color = 'b', alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));
```

```python
# Plot the histogram for BTC and ETH returns using 50 bins. Display all histograms on the same figure
plt.figure(figsize = (20, 15)) 

mu_BTC = round(daily_return_df['BTC'].mean(), 2)
sigma_BTC = round(daily_return_df['BTC'].std(), 2)
 

mu_ETH = round(daily_return_df['ETH'].mean(), 2)
sigma_ETH = round(daily_return_df['ETH'].std(), 2)

daily_return_df[['ETH','BTC']].plot.hist(bins = 60, alpha = 0.7, figsize = (15, 8))
plt.grid()
plt.ylabel('Probability')

plt.title('Histogram: BTC mu=' + str(mu_BTC) + ', BTC sigma=' +str(sigma_BTC) + ', ETH mu=' +str(mu_ETH) + ', ETH sigma=' +str(sigma_ETH) );
```
