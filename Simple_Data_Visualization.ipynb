```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# Define a Pandas Dataframe with all crypto allocation in the portfolio
crypto_df = pd.DataFrame(data = {'allocation %':[20, 45, 5, 12, 3, 10, 5]}, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA', 'AXS', 'SOL'])
crypto_df
```

```python
# Use matplotlib to plot a pie chart 
crypto_df.plot.pie(y = 'allocation %', figsize = (8, 8))
plt.title('CRYPTO PORTFOLIO PIE CHART');
```

```python
# Use 'explode' to increase the separation between ETH and the rest of the portfolio
explode = (0, 0.2, 0, 0, 0, 0, 0)
crypto_df.plot.pie(y = 'allocation %', figsize = (8, 8), explode = explode)
plt.title('CRYPTO PORTFOLIO PIE CHART');
```
